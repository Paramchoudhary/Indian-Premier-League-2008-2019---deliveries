## Boxplot Method to spot the Outliers

```python
sns.boxplot(df['batsman_runs'])
```

```python
cols = 5
rows = 5
num_cols = df.select_dtypes(exclude='object').columns
fig = plt.figure(figsize= (15,9))
for i, col in enumerate(num_cols):
    
    ax=fig.add_subplot( rows, cols, i+1)
    
    sns.boxplot(x=df[col], ax=ax)
    
fig.tight_layout()  
plt.show()


#Here, the numerical columns must be selected in order to visualise the histplot. For that, Datatypes like 'Object'
#(i.e, categorical) need to be excluded from the process of importing.
# Note: 'plt.show()' is a mandatory command in order to print the executed process. 
```

From the above method of spotting the outliers for all columns together, it can be clearly inferred that OUTLIERS are present in the columns [match_id], [ball], and [inning]. Rest of the columns DO NOT contain any outliers.

## Method of removing outliers from the visualised boxplot:

Sometimes a dataset can contain extreme values that are outside the range of what is expected and unlike the other data. These are called outliers and often machine learning modeling and model skill in general can be improved by understanding and even removing these outlier values.

This does not mean that the values identified are outliers and should be removed.

For example, If we consider the column "match_id", there are n number of outliers. But, if we observe the column "balls" refers to "The innings played by a sinle player". 

So, Removing Outliers all the time is not a good method.

## Spotting the outliers using Z-Score method:

```python
from scipy import stats
import numpy as np
 
z = np.abs(stats.zscore(df['over']))
print(z)
```

## Spotting the Outliers using IQR method:

```python
Q1 = np.percentile(df['over'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(df['over'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
# Above Upper bound
upper = df['over'] >= (Q3+1.5*IQR)
 
print("Upper bound:",upper)
print(np.where(upper))
 
# Below Lower bound
lower = df['over'] <= (Q1-1.5*IQR)
print("Lower bound:", lower)
print(np.where(lower))
```

## Removing the Outliers after the whole analysis.

```python
Q1 = np.percentile(df['batsman_runs'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(df['batsman_runs'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
 
print("Old Shape: ", df.shape)
 
# Upper bound
upper = np.where(df['batsman_runs'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(df['batsman_runs'] <= (Q1-1.5*IQR))
 
''' Removing the Outliers '''
df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)
 
print("New Shape: ", df.shape)
```
